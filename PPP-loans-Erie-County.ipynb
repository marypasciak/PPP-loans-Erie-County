```python
# extract only the loans for projects located in NYS

df_relevant[df_relevant.ProjectState == 'NY']
df_nys = df_relevant[df_relevant.ProjectState == 'NY']
df_nys.head(2)
```

```python
# extract only the loans with a ProjectCountyName of ERIE
# call this df_erie

df_erie = df_nys[df_nys.ProjectCountyName == 'ERIE']
df_erie.head(2)
```

```python
# export df_erie to a .csv file

df_erie.to_csv(r'df_erie.csv')
```

```python
# how many different NAICSCodes are there in Erie County?

unique_value = df_erie["NAICSCode"].nunique()
print(unique_value)
```

```python
# How many loans were granted to each NAICSCode in Erie County?

df_loans_per_NAICS_erie = df_erie.groupby('NAICSCode').NAICSCode.count()
df_loans_per_NAICS_erie.head(2)
```

```python
# create a .csv file of the number of loans granted to each NAICSCode in Erie County
# df.to_csv(r'Path where you want to store the exported CSV file\File Name.csv', index = False)

df_loans_per_NAICS_erie.to_csv(r'df_loans_per_NAICS_erie.csv')
```

```python
# Aggregate multiple columns at once

# Total number of loans to each NAICSCode in Erie County? count
# Total number of JobsReported to each NAICSCode in Erie County? sum
# Total InitialApprovalAmount loaned to each NAICSCode in Erie County? sum
# Total ForgivenessAmount to each NAICSCode in Erie County? sum

grouped_df_erie = df_erie.groupby("NAICSCode")

function_dict = {
    "ProjectCountyName": "count",
    "JobsReported": "sum",
    "CurrentApprovalAmount": "sum",
    "ForgivenessAmount": "sum",
    "PAYROLL_PROCEED": "sum"
}

grouped_df_erie = grouped_df_erie.aggregate(function_dict)

print(grouped_df_erie)
```

```python
# export to .csv file

grouped_df_erie.to_csv(r'grouped_df_erie.csv')
```

```python
# read in naics-codes.csv

#df = pd.read_csv("public_150k_plus.csv")
#df

df_naics_codes = pd.read_csv("naics-codes.csv")
df_naics_codes.head(2)
```

```python
# create a row at the top of df_naics_codes for column headers

df_naics_codes.columns = ['NAICSCode','business_type'] 
df_naics_codes.head(3)
```

```python
df_naics_codes.dtypes
```

```python
# in original df, the datatype of NAICSCode was float64

# need to change the datatype of NAICS_code column in df_naics_codes to float
#    so that it matches the datatype of  NAICSCode column in df_erie

df_naics_codes.NAICSCode = df_naics_codes.NAICSCode.astype(float)
df_naics_codes.dtypes
```

```python
# join df_erie_aggregated on NAICSCode to df_naics_codes on NAICSCode 
# create a new df of this called df_erie_naics_agg

df_erie_naics_agg=pd.merge(
    grouped_df_erie,
    df_naics_codes,
    on='NAICSCode')

df_erie_naics_agg
```

```python
# export to a .csv file

df_erie_naics_agg.to_csv(r'df_erie_naics_agg.csv')
```

```python
# Which business types in Erie County received the most loans?

df_erie_naics_agg.sort_values(by='ProjectCountyName',ascending=False).head(10)
```

```python
# Which business types in Erie County received the largest total CurrentApprovalAmount?

df_erie_naics_agg.sort_values(by='CurrentApprovalAmount',ascending=False).head(10)
```

```python
# Which business types in Erie County received money for the most JobsReported?

df_erie_naics_agg.sort_values(by='JobsReported',ascending=False).head(10)
```

```python
# Which business types in Erie County had the greatest percent of their CurrentApprovalAmount forgiven?

# add a column that calculates, for each business type,
# the percent of money loaned that has been forgiven
# ForgivenessAmount / CurrentApprovalAmount

df_erie_naics_agg['PctOfLoanForgiven'] = df_erie_naics_agg.apply(lambda row: ((row.ForgivenessAmount / 
                                  row.CurrentApprovalAmount) * 100), axis = 1).round()

df_erie_naics_agg.sort_values(by='PctOfLoanForgiven',ascending=False).head(10)
```

```python
# Which business types in Erie County spent the most money per job?

# add a column that calculates, for each business type,
# the average amount of money from PAYROLL_PROCEED spent per job
# PAYROLL_PROCEED / JobsReported

df_erie_naics_agg['AvgPayrollSpentPerJob'] = df_erie_naics_agg.apply(lambda row: (row.PAYROLL_PROCEED / 
                                  row.JobsReported), axis = 1).round()

df_erie_naics_agg.sort_values(by='AvgPayrollSpentPerJob',ascending=False).head(10)
```

```python
# export df_erie_naics_agg -- with the new calculated columns -- as a .csv

df_erie_naics_agg.to_csv(r'erie_business_types_aggregated.csv')
```
